Removing columns with more than 70% "0" entries

```
totalEntries = trialDataset.shape[0]
for column in trialDataset.iloc[0:, 4:]:
  if (trialDataset[column].isin([0]).sum() / totalEntries) * 100 >= 70:
        trialDataset.drop(column,axis=1, inplace=True)

cleanedGroupedData = trialDataset.groupby("Country/Region").sum()
```
